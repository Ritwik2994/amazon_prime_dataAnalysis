```python
#Importing the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the dataset
prime = pd.read_csv("prime_video_dataset.csv",encoding="iso-8859-1")
```

## Exploratory Analysis and Visualization

Visualization in data compution is the key to achieve the result in fast and efficiect manner and it also very helpful in analysing the data just buy going into the graphs. 

In this section I have used seaborn and matplotlib libraries to define the data into a visualized form. 

functions which I've used
 1) plt.figure ==> for ploting graph size
 
 2) sns.counterplot ==> for ploting graph
 
 3) plt.show ==> for displaying the graph
 
They all acts as the basic for graphing the data. just we have defing what kind of graph we want. In this section I have used bar graph, scatter graph etc.


```python
#checking language of the contents available in the datset
p_df['Language'].value_counts()
```

```python
#displaying the above information in a bar graph form
plt.figure(figsize=(15,10))
sns.countplot(x=p_df['Language'], edgecolor='black');
plt.show()
```

```python
#discribing initial statistic
p_df.describe()
```

```python
#displaying Genre of the video available in Prime Video
plt.figure(figsize=(15,10))
sns.countplot(y=p_df['Genre'], edgecolor='black')
plt.show()
```

```python
#ploting graph using above informations 
plt.figure(figsize=(15, 20))
plt.xlabel("Language")
plt.ylabel("Genre")
sns.scatterplot(y=p_df['Genre'],x=p_df['Language']);
```

```python
#ploting scatter graph 
plt.figure(figsize=(15,25))
plt.xlabel("Number of Seasons")
plt.ylabel("Genre")
plt.scatter(p_df['No of seasons available'],p_df['Genre'])
plt.grid(linestyle='dotted')
plt.show()
```

Q2. What are the top 10 genres ? Display it visually

```python
p_df["Genre"].value_counts().head(10)
```

```python
#Visual representation of top 10 geners
#displaying the above information in a bar graph form
plt.figure(figsize=(20,8))
p_df["Genre"].value_counts()[:10].plot(kind="barh")
plt.title("Top 10 Genres",size=20)
```

Q3. Graphically calculate different viewership age with respect to contents.

```python
#importing counter from collection
from collections import Counter

#calculating counter variable
size = Counter(p_df['Age of viewers']).values()
labels = Counter(p_df['Age of viewers']).keys()

#ploting Graph
plt.figure(figsize=(8,8))
plt.title("Viewership Age with respect to contents", fontweight = "bold")
plt.pie(size, labels=labels, autopct='%1.1f%%',startangle=140)
plt.axis('equal')
plt.show()
```
